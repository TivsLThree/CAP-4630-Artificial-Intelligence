```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### f1 with minimum at (0,0,0)

```
def f(x, y):
  return  x**2 + y**2
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-150, 150, 80)
y = np.linspace(-150, 150, 80)

ax.scatter3D(0,0,0)
ax.plot3D(x,y,f(x,y), 'gray')
```

```
def f(x, y):
  return  x**2 + y**2
fig = plt.figure()
x = np.linspace(-150, 150, 40)
y = np.linspace(-150, 150, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
contours = plt.contour(X, Y, Z, 50,cmap='RdGy');
plt.clabel(contours, fontsize = 0);
plt.imshow(Z, extent=[-150,150, -150, 150], origin='lower', cmap='RdGy', alpha = 0.9)
```
